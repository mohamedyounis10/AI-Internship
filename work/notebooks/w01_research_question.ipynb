## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Unit of Analysis: A single content item / page (content_id).

Decision: Identifying which content characteristics and signals are most strongly correlated with high organic visibility and clicks, so we can prioritize optimization efforts.

Action: Content strategists can update content creation guidelines (e.g., minimum depth/word count, refresh intervals, or targeting optimal position tiers) based on data-backed signals.

Cost of a Wrong Call: Misidentifying a signal could lead to wasting editorial time rewriting pages based on irrelevant metrics, or misallocating resources away from factors that actually drive search traffic.

Why Data/ML Helps: With thousands of pages, manual inspection of ranking factors is impossible. Statistical analysis and ML feature importance allow us to evaluate multiple signals simultaneously at scale.
